```python
def timingfromhours(x):
    if x in range(4, 12):
        timing = 'Morning'
    elif x in range(12, 17):
        timing = 'Afternoon'
    elif x in range(17, 22):
        timing = 'Evening'
    elif x in [22,23,0,1,2,3]:
        timing = 'Night'
    else:
        timing = 'X'
    return timing
```

```python
#Grouping Fields by Month
def month_grouping(col):
    return dataf[[col,'month']].groupby('month').agg({col:['mean']})[col]
```

```python
#Grouping Fields by Weekday
def Weekday_grouping(col):
    Weekdays = dataf.groupby('weekday').agg({col:['mean']})
    Weekdays.columns = [f"{i[0]}_{i[1]}" for i in Weekdays.columns]
    Weekdays['week_num'] = [['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'].index(i) for i in Weekdays.index]
    Weekdays.sort_values('week_num', inplace=True)
    Weekdays.drop('week_num', axis=1, inplace=True)
    return Weekdays
```

```python
def time_grouping(col):
    timings = dataf.groupby('timing').agg({col:['mean']})
    timings.columns = [f"{i[0]}_{i[1]}" for i in timings.columns]
    timings['timing_num'] = [['Morning','Afternoon','Evening','Night'].index(i) for i in timings.index]
    timings.sort_values('timing_num', inplace=True)
    timings.drop('timing_num', axis=1, inplace=True)
    return timings
```

```python
# Change Detection Model for predicting change in Energy Consumption 

def chng_detection(col_1, _r=0.01, _order=1, _smooth=10):
    cf = changefinder.ChangeFinder(r=_r, order=_order, smooth=_smooth)
    change_df = pd.DataFrame()
    col = 'House Overall'
    change_df[col] = dataf[col_1].resample('D').mean()
    # calculate the change score
    change_df['change_score'] = [cf.update(i) for i in change_df[col]]
    change_score_q1 = stats.scoreatpercentile(change_df['change_score'], 25)
    change_score_q3 = stats.scoreatpercentile(change_df['change_score'], 75)
    thr_upper = change_score_q3 + (change_score_q3 - change_score_q1) * 3

    anom_score = hv.Curve(change_df['change_score'])
    anom_score_th = hv.HLine(thr_upper).opts(color='red', line_dash="dotdash")

    anom_points = [[change_df.index[a],change_df[col][a]] for a, score in enumerate(change_df["change_score"]) if score > thr_upper]
    org = hv.Curve(change_df[col],label='House Overall').opts(yformatter='%.1fkw')
    detected = hv.Points(anom_points, label=f"House Overall detected").opts(color='red', legend_position='bottom', size=5)

    return ((anom_score * anom_score_th).opts(title=f"House Overall Change Score & Threshold") + \
            (org * detected).opts(title=f"House Overall Detected Points")).opts(opts.Curve(width=800, height=300, show_grid=True, tools=['hover'])).cols(1)
```

```python
#Rescale
data_daily_lstm = dataf[['use_HO', 'Furnace', 'Living room', 'Barn', 'temperature', 'humidity',
                   'apparentTemperature', 'pressure', 'cloudCover','windBearing', 'precipIntensity',
                   'dewPoint', 'precipProbability']]
data_daily = data_daily_lstm.resample('D').mean()
#Normalize the features
scaler = MinMaxScaler(feature_range=(0, 1))
data_daily[data_daily.columns[1:]] = scaler.fit_transform(data_daily[data_daily.columns[1:]])
scaler_target = MinMaxScaler(feature_range=(0, 1))
data_daily[['use_HO']] = scaler_target.fit_transform(data_daily[['use_HO']])

size = int(len(data_daily)*0.7)
data_daily_train = data_daily[:size]
data_daily_test = data_daily[size:]
X_train, X_test = [], []
Y_train, Y_test = [], []
n_past=1
n_future=1
for i in range(n_past, len(data_daily_train)-n_future+1):
    X_train.append(data_daily_train.iloc[i-n_past:i, 0:data_daily.shape[1]])
    Y_train.append(data_daily_train.iloc[i+n_future-1:i+n_future, 0])
for i in range(n_past, len(data_daily_test)-n_future+1):
    X_test.append(data_daily_test.iloc[i-n_past:i, 0:data_daily_test.shape[1]])
    Y_test.append(data_daily_test.iloc[i+n_future-1:i+n_future, 0])
    
X_train, Y_train = np.array(X_train), np.array(Y_train)
X_test, Y_test = np.array(X_test), np.array(Y_test)

print('X_train shape', X_train.shape)
print('X_test shape', X_test.shape)
print('Y_train shape', Y_train.shape)
print('Y_test shape', Y_test.shape)
```

```python
model = Sequential()
model.add(LSTM(25, activation='relu', return_sequences = False, input_shape=(X_train.shape[1], X_train.shape[2])))
#model.add(LSTM(50, activation='relu', return_sequences = True))
#model.add(LSTM(15, activation='relu', return_sequences = False))
#model.add(Dropout(0.2))
model.add(Dense(Y_train.shape[1]))
model.compile(optimizer='adam', loss='mse')
model.summary()
```

```python
model_fit = model.fit(X_train, Y_train, epochs=60, verbose=0)

Train_pred = model.predict(X_train, verbose=0)
Y_pred = model.predict(X_test, verbose=0)
```

```python
plt.plot(model_fit.history['loss'], label='Train loss')
plt.legend()
print('Train MSE minimum:', min(model_fit.history['loss']))
```

```python
data_daily[['use_HO']] = scaler_target.inverse_transform(data_daily[['use_HO']])
Y_pred = scaler_target.inverse_transform(Y_pred)
Train_pred = scaler_target.inverse_transform(Train_pred)
```
